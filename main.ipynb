```python
# using RK4 for the solution of Lane-Emden
# last edited in 2023/5/12  By Wan
import numpy as np
import matplotlib.pyplot as plt
import RungeKutta as RK4
```

```python
# Set basic paras, including n and step length
n = 1.0; step_len = 0.05
```

```python
# calculate the result
def RK_cal(n, step_len):
#set initial values. The value of x0 should be small enough, but not zero
     x0 = 1e-100
     y0 = 1.0
     z0 = 0
     arr1 = [x0]
     arr2 = [y0]
     if n==2 or n==0: # when n=0 or n=2, the value of the solutions will decrease rapidly
          while y0 > -0.25:
               x0, y0, z0 = RK4.RK_solve(x0, y0, z0, n, step_len)
               arr1.append(x0)
               arr2.append(y0)
          return arr1, arr2
     else:
          while x0 < 20:
               x0, y0, z0 = RK4.RK_solve(x0, y0, z0, n, step_len)
               arr1.append(x0)
               arr2.append(y0)
          return arr1, arr2
```

```python
# calculate n = 1 condition
if __name__ == '__main__':
    xs1, ys1 = RK_cal(n, step_len)
    plt.plot(xs1, ys1, label = 'runge-kutta')
    plt.show()
```

```python
# calculate solutions from n=0 to 6 
xs0, ys0 = RK_cal(0, step_len)
xs1, ys1 = RK_cal(1.0, step_len)
xs2, ys2 = RK_cal(2.0, step_len)
xs3, ys3 = RK_cal(3.0, step_len)
xs4, ys4 = RK_cal(4.0, step_len)
xs5, ys5 = RK_cal(5.0, step_len)
xs6, ys6 = RK_cal(6.0, step_len)
xs7, ys7 = RK_cal(1.5, step_len)


# plot all n values 
plt.figure(dpi=128)
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
ax = plt.gca()
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.plot(xs0, ys0, label = 'n=0')
plt.plot(xs1, ys1, label = 'n=1')
plt.plot(xs7, ys7, label = 'n=1.5')
plt.plot(xs2, ys2, label = 'n=2')
plt.plot(xs3, ys3, label = 'n=3')
plt.plot(xs4, ys4, label = 'n=4')
plt.plot(xs5, ys5, label = 'n=5')
plt.plot(xs6, ys6, label = 'n=6')
plt.legend()
```

```python
# compare the numerical solutions with the analytical solutions (only when n=0,1,5)
plt.figure(dpi=128)
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
ax = plt.gca()
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
x=np.linspace(0,2.75,20000)
y = 1-(1/6)*x**2
plt.plot(x, y, label = 'analytical')
plt.plot(xs0, ys0, label = 'numerical')
plt.legend()

plt.figure(dpi=128)
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
ax = plt.gca()
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
x=np.linspace(0,20,20000)
np.seterr(divide='ignore',invalid='ignore')
y = np.sin(x)/x
plt.plot(x, y, label = 'analytical')
plt.plot(xs1, ys1, label = 'numerical')
plt.legend()

plt.figure(dpi=128)
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
ax = plt.gca()
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
x=np.linspace(0,20,20000)
y = ((1+(x**2/3))**-0.5)
plt.plot(x, y, label = 'analytical')
plt.plot(xs5, ys5, label = 'numerical')
plt.legend()
```
